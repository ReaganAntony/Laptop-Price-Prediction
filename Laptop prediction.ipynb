```python
from sklearn.preprocessing import LabelEncoder
```

```python
encoder=LabelEncoder()
```

```python
df1=data.select_dtypes(include="object")
df2=data.select_dtypes(exclude="object")
```

```python
df1=df1.select_dtypes(include="object").apply(encoder.fit_transform)
```

```python
df1
```

```python
data=pd.concat((df1,df2),axis=1)
```

```python
data
```

```python
data["Touch_screen"]=data["Touch_screen"].astype(int)
```

```python
data.sample(3)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x=data.drop(["Price"],axis=1)
y=data["Price"]
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
x = scaler.fit_transform(x)
x = pd.DataFrame(x)
```

```python
x
```

```python
y
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,random_state=10, train_size=0.8) 
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
lr.score(x_train,y_train)
```

```python
lr.score(x_test,y_test)
```

```python
from sklearn.svm import SVR
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.neighbors import KNeighborsRegressor
```

```python
lis=[SVR,DecisionTreeRegressor,RandomForestRegressor,KNeighborsRegressor]
```

```python
for i in lis:
    model = i()
    model.fit(x_train,y_train)
    print(i)
    print(model.score(x_train,y_train))
    print(model.score(x_test,y_test))
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
rf = RandomForestRegressor(random_state=20)

# Define hyperparameters grid
param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'max_features': ['auto', 'sqrt', 'log2']
}

# Grid Search
grid_search = GridSearchCV(estimator=rf, param_grid=param_grid, cv=5, scoring='neg_mean_squared_error', verbose=2)
grid_search.fit(x_train, y_train)

# Best hyperparameters
print(f'Best Hyperparameters: {grid_search.best_params_}')
```
